# 1. Autocorrelation coefficients

Autocorrelation is the correlation between a variable lagged one or
more time periods and itself.

Q1.Use Excel to compute the autocorrelation coefficients and display the autocorrelation function for the data presented in Table 1.

```python
import pandas as pd

df = pd.DataFrame (
    {
        "Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
        "VCR":[123,130,125,138,145,142,141,146,147,157,150,160]
    }
)

print(df)
```

![title](img\img1.png)

```python
df["Yt_1"]=df["VCR"].shift(1)
df["Yt_2"]=df["VCR"].shift(2)
df["Yt_3"]=df["VCR"].shift(3)
df
```

```python
# mean Yt_k.

Ybar = df["VCR"].mean()
sum_Yt = df["VCR"].sum()
mean_Yt_1 = df["Yt_1"].mean()
mean_Yt_2 = df["Yt_2"].mean()
mean_Yt_3 = df["Yt_3"].mean()

print(sum_Yt)
print(Ybar)
print(mean_Yt_1)
print(mean_Yt_2)
print(mean_Yt_3)
```

```python
df["Yt-Ybar"] = df["VCR"]-Ybar
df["Yt_1-Ybar"] = df["Yt_1"]-Ybar
df["Yt_2-Ybar"] = df["Yt_2"]-Ybar
df["Yt_3-Ybar"] = df["Yt_3"]-Ybar

df
```

```python
df["(Yt-Ybar)(Yt_1-Ybar)"]=df["Yt-Ybar"]*df["Yt_1-Ybar"]
df["(Yt-Ybar)(Yt_2-Ybar)"]=df["Yt-Ybar"]*df["Yt_2-Ybar"]
df["(Yt-Ybar)(Yt_3-Ybar)"]=df["Yt-Ybar"]*df["Yt_3-Ybar"]
df["(Yt-Ybar)^2"]=df["Yt-Ybar"]**2
df
```

```python
r1=df["(Yt-Ybar)(Yt_1-Ybar)"].sum()/df["(Yt-Ybar)^2"].sum()
r2=df["(Yt-Ybar)(Yt_2-Ybar)"].sum()/df["(Yt-Ybar)^2"].sum()
r3=df["(Yt-Ybar)(Yt_3-Ybar)"].sum()/df["(Yt-Ybar)^2"].sum()
print(r1)
print(r2)
print(r3)
```

### Results 
Positive lag 1 autocorrelation exists in this time series. The correlation between Yt and Yt-1
or the autocorrelation for time lag 1, is .572. This means that the successive monthly sales of VCRs are somewhat correlated with each other.

It appears that moderate autocorrelation exists in this time series lagged two time periods. The correlation between Yt and Yt-2
, or the autocorrelation for time lag 2, is .463.

The autocorrelation for time lag 3, is .111. This means a weak autocorrelation exists in this
time series lagged three time periods.

Notice that the autocorrelation coefficient at time lag 2 (.463) is less than the
autocorrelation coefficient at time lag 1 (.572). Generally, as the number of time lags ( k )
increases, the magnitudes of the autocorrelation coefficients decrease.

# 2. Are the Data Random?

- If a series is random, the autocorrelations between Yt
and Yt-k
for any
time lag k are close to zero. The successive values of a time series are
not related to each other.


## 2.1 To determine whether an autocorrelation coefficient is significantly different from zero

- Compute the standard deviations (or standard errors) of the autocorrelation coefficients.

- Compute the Ljung-Box Q statistics to examine a set of consecutive rk’s as a group.

## 2.2 Do the Data have a Trend?
- If a series has a trend, successive observations are highly correlated,
and the auto- correlation coefficients typically are significantly
different from zero for the first several time lags and then gradually
drop toward zero as the number of lags increases. The
autocorrelation coefficient for time lag 1 will often be very large
(close to 1). The autocorrelation coefficient for time lag 2 will also be
large. However, it will not be as large as for time lag 1.

## 2.3 Are the Data Seasonal?
- If a series has a seasonal pattern, a significant autocorrelation
coefficient will occur at the seasonal time lag or multiples of the
seasonal lag. The seasonal lag is 4 for quarterly data and 12 for
monthly data.

Problem 1

An analyst would like to determine whether there is a pattern to earnings per share for the Price Company, which operated a wholesale/retail cash-and-carry business in several states under the name Price Club. The data are shown in Table P-20. (You may download the data file from Moodle)

a) i) Describe any patterns that exist in these data.

ii) Find the forecast value of the quarterly earnings per share for Price Club

for each quarter by using the naive approach (i.e., the forecast for firstquarter 1987 is the value for fourth-quarter 1986, .32).

b) Evaluate the naive forecast using MAD.

c) Evaluate the naive forecast using MSE and RMSE.

d) Evaluate the naive forecast using MAPE.

e) Evaluate the naive forecast using MPE. 

```python
df = pd.DataFrame(
    {
    "Yt":[0.4,0.29,0.24,0.32,0.47,0.34,0.3,0.39,0.63,0.43,0.38,0.49,0.76,0.51,0.42,0.61,0.86,0.51,0.47,0.63,0.94,0.56,0.5,0.65,0.95,0.42,0.57,0.6,0.93,0.38,0.37,0.57
]
    }
)
df
```

```python
# focast value
df["YtCap"] = df["Yt"].shift(1)
# error
df["Yt-YtCap"] = df["Yt"]-df["YtCap"] 
#abs error
df["abs(Yt-YtCap)"] = df["Yt-YtCap"].abs()
#error ^ 2
df["(Yt-YtCap)^2"] = df["Yt-YtCap"]**2
#abs error /abs true value
df["abs(Yt-YtCap)/abs(Yt)"] = df["abs(Yt-YtCap)"]/df["Yt"].abs()
#error / true value
df["(Yt-YtCap)/Yt"] = df["Yt-YtCap"]/df["Yt"]
df
```

```python
#mean absolute deviation (abs average error  )
MAD = df["abs(Yt-YtCap)"].sum()/df["YtCap"].count()
#mean squared error
MSE = df["(Yt-YtCap)^2"].sum()/df["YtCap"].count()
#root mean squared error
RMSE = MSE ** (1/2)
#mean absolute percentage error
MAPE = df["abs(Yt-YtCap)/abs(Yt)"].sum()/df["YtCap"].count()
#mean percentage error
MPE = df["(Yt-YtCap)/Yt"].sum()/df["YtCap"].count()
print(MAD , MSE , RMSE , MAPE , MPE)
```
